```python
import pandas as pd
#######CLEANING DATASET###########
# Load the dataset
df = pd.read_csv("netflix_titles.csv")

print("=" * 50)
print("Original Dataset Shape:")
print(df.shape)

print("\n" + "=" * 50)
print("Missing Values:")
print(df.isnull().sum())

print("\n" + "=" * 50)
print("Duplicate Rows:")
print(df.duplicated().sum())

# Remove duplicate rows
df = df.drop_duplicates()

print("\n" + "=" * 50)
print("Shape After Removing Duplicates:")
print(df.shape)

# Fill missing values
df["director"] = df["director"].fillna("Unknown")
df["cast"] = df["cast"].fillna("Unknown")
df["country"] = df["country"].fillna("Unknown")
df["rating"] = df["rating"].fillna("Not Rated")
df["duration"]=df["duration"].fillna("unknown")



# Remove extra spaces
df["date_added"] = df["date_added"].str.strip()

# Convert to datetime
df["date_added"] = pd.to_datetime(df["date_added"], errors="coerce")
print("\n" + "=" * 50)
print("Data Types:")
print(df.dtypes)

print("\n" + "=" * 50)
print("Remaining Missing Values:")
print(df.isnull().sum())



#####################
# ============================
# Movies vs TV Shows
# ============================

print("\n" + "=" * 50)
print("Movies vs TV Shows")

content_count = df["type"].value_counts()

print(content_count)

#############BAR CHART: MOVIES AND SHOWS
import matplotlib.pyplot as plt
content_count.plot(kind="bar")
plt.title("Movies vs TV Shows on Netflix")
plt.xlabel("Content Type")
plt.ylabel("Number of Titles")

plt.show()


```

```python
######
# ============================
# Top 10 Countries
# ============================

top_countries = df["country"].value_counts().head(10)

print("\nTop 10 Countries:")
print(top_countries)

top_countries.plot(kind="bar", figsize=(10,5))

plt.title("Top 10 Countries with Netflix Content")
plt.xlabel("Country")
plt.ylabel("Number of Titles")

plt.xticks(rotation=45)

plt.show()
```

```python
# ============================
# Content Released by Year
# ============================

release_year = df["release_year"].value_counts().sort_index()

print("\nContent Released by Year:")
print(release_year.tail(10))
```

```python
release_year.plot(kind="line",figsize=(12,6))

plt.title("netflix content released by Year")
plt.xlabel("release year")
plt.ylabel("number of Titles")
plt.show()
```

```python
##########
###top 10 genres

top_genres = df["listed_in"].value_counts().head(10)

print("top 10 genres")
print(top_genres)
```

```python

top_genres.plot(kind="barh",color="skyblue",figsize=(10,6))
plt.title("top 10 genres on netflix") 
plt.xlabel("number of titles")
plt.ylabel("genres")

plt.show()
```

```python
# ============================
# Content Ratings Analysis
# ============================
rating=df["rating"].value_counts()
print("content rating")
print(rating)
```

```python
rating.plot(kind="bar",color="darkblue",figsize=(10,6))
plt.title("Netflix Content Rating")
plt.xlabel("Rating")
plt.ylabel("Number of Titles")

plt.xticks(rotation=45)
plt.show()
```

```python

#####movies added every year 
added_year=df["date_added"].dt.year.value_counts().sort_index()
print(added_year)

added_year.plot(kind="line",color="green",figsize=(10,5))

plt.title("content added to netflix by year")
plt.xlabel("year")
plt.ylabel("number of titles")
plt.grid(True)
plt.show()
```
